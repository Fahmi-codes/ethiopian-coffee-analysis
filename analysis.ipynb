```python
import matplotlib.pyplot as plt

df["Country.of.Origin"].value_counts().head(10).plot(kind="bar")

plt.title("Top Coffee Producing Countries in Dataset")
plt.xlabel("Country")
plt.ylabel("Count")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()

sns.histplot(df["Total.Cup.Points"], bins=30)

plt.title("Distribution of Coffee Quality Scores")
plt.xlabel("Total Cup Points")
plt.ylabel("Frequency")

plt.show()
```

```python
country_scores = df.groupby("Country.of.Origin")["Total.Cup.Points"].mean().sort_values(ascending=False).head(10)

plt.figure()

country_scores.plot(kind="bar")

plt.title("Top Countries by Average Coffee Score")
plt.xlabel("Country")
plt.ylabel("Average Cup Points")

plt.show()
```

```python
plt.figure()

df["Processing.Method"].value_counts().head(8).plot(kind="bar")

plt.title("Coffee Processing Methods")
plt.xlabel("Processing Method")
plt.ylabel("Count")

plt.show()
```
